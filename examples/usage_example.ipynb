```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

```python
n_samples = 300
n_features = 2
random_state = 42

#cluster data
X_clusters, _ = make_blobs(n_samples=n_samples, centers=3, n_features=n_features, cluster_std=0.5, random_state=random_state)

#noise data
rng = np.random.RandomState(random_state)
X_noise = rng.uniform(low=-9, high=8, size=(60, n_features))

data_values = np.vstack((X_clusters, X_noise))

#clustering
db = DBSCAN(eps=0.7, min_samples=5)
labels = db.fit_predict(data_values)

plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = [plt.cm.tab10(i / 10) for i in range(len(unique_labels))]
for label, color in zip(unique_labels, colors):
    mask = labels == label
    label_name = f"Cluster {label}" if label != -1 else "Noise -1"
    plt.scatter(data_values[mask, 0], data_values[mask, 1], s=50, c=[color], label=label_name)
plt.legend()
plt.title("DBSCAN Cluster with Noise")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
from nacvi.daviesbouldin_plus import dbi_plus_score

print("dbi+ score:", dbi_plus_score(data_values, labels))
```
